```python
# Import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts size of plots created


# Read in the data

df = pd.read_csv(r'C:\Users\admin\Downloads\movies.csv')
```

```python
df.head()
```

```python
# Searching for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
# Removing missing data

df = df.dropna() 
```

```python
# Determining data types for each column

df.dtypes
```

```python
# Changing datatype of 'Budget','Gross', and 'Votes' columns to integer

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df['votes'] = df['votes'].astype('int64')
```

```python
df.head()
```

```python
# Creating corrected Year column from Released column

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)

df.head()
```

```python
# Sorting table by Gross descending

df = df.sort_values(by=['gross'],inplace=False, ascending=False)
```

```python
# Setting option to display all available rows in table

pd.set_option('display.max_rows', None)
```

```python
# Showing sample of unique production companies from Company column

df['company'].drop_duplicates().sort_values(ascending=False)[:5]

```

```python
# Creating a scatterplot showing Budget vs Gross

plt.scatter(x=df['budget'],y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget')

plt.show()
```

```python
df.head()
```

```python
# Plotting Budget vs Gross using Seaborn to better show correlation

sns.regplot(x='budget',y='gross', data=df, scatter_kws={"color":"blue"},line_kws={"color":"orange"})
```

```python
# Examining correlation

df.corr(method='pearson') 
```

```python
# Visualizing correlation with a heatmap matrix

correlation_matrix = df.corr(method='pearson') 

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Converting object type data into numerical data to use in heatmap matrix

df_numerized = df.copy()

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
df.head()
```

```python
# Updating heatmap matrix to include all values

correlation_matrix = df_numerized.corr(method='pearson') 

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Examining correlation further

df_numerized.corr()
```

```python
# Creating correlated pairs

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs.head()
```

```python
# Sorting correlated pairs

sorted_pairs = corr_pairs.sort_values()

sorted_pairs.head()
```

```python
# Filtering out variables with low correlation

high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
